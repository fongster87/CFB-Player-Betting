```python
passing_box = pd.read_excel("CFB Box Scores.xlsx", sheet_name= "Passing")
passing_box["C/ATT"] = passing_box["C/ATT"].str.strip("/") 
passing_box[["Completions", "Attempts"]] = passing_box["C/ATT"].str.split("/", expand = True)
passing_box["Completions"] = pd.to_numeric(passing_box["Completions"])
passing_box["Attempts"] = pd.to_numeric(passing_box["Attempts"])
passing_box
```

```python
defense_stats = pd.read_excel("CFB Model Data.xlsx", sheet_name= "Team Defense")
defense_stats
```

```python
offense_stats = pd.read_excel("CFB Model Data.xlsx", sheet_name= "Team Offense")
offense_stats
```

```python
qb_stats = pd.read_excel("CFB Player Stats.xlsx", sheet_name="Passing Model Data")
qb_stats
```

```python
all_pass = passing_box.merge(qb_stats, on = "Player")
all_pass = all_pass.merge(offense_stats, on = "Team")
all_pass = all_pass.merge(defense_stats, on = "Opponent")
all_pass
```

```python
home_away_map = {
    "Home" : 1, 
    "Away" : 0, 
    "Neutral Site": 2
}
all_pass["Home/Away"] = all_pass["Home/Away"].map(home_away_map)
all_pass.to_excel("a.xlsx")
```

```python
passing = pd.read_excel("CFB Model Data.xlsx", sheet_name = "Passing", skiprows=1)
passing.to_csv("passing.csv")
```

```python
rush_box =  pd.read_excel("CFB Box Scores.xlsx", sheet_name= "Rushing")
rush_box["Player"] = rush_box["Player"].str.replace(r"#\d+", "", regex=True).str.strip()
rush_box
```
